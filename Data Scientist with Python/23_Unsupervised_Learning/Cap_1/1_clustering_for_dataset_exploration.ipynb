```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import make_pipeline
import pandas as pd
import os
from datetime import datetime
import plotly.graph_objs as go
```

```python
def plot_price(df, column='Equity'):
    df[column].plot(figsize=(10, 6))
    plt.title('Equity Curve')
    plt.xlabel('Date')
    plt.ylabel('Equity Value')
    plt.grid(False)
    plt.show()
    
```

```python

br_directory_path = '/home/jm/Projetos/Pessoais/stock-market/stock-market-scraper/py_scraper_scripts/csv/prices/BR'
br_files_list = [f for f in os.listdir(br_directory_path) if os.path.isfile(os.path.join(br_directory_path, f))]


start_date = '2022-09-17'
end_date = f"{datetime.now().year}-{datetime.now().month}-{datetime.now().day}"

initial_equity = 1
df_bench = pd.read_csv('/home/jm/Projetos/Pessoais/stock-market/stock-market-scraper/py_scraper_scripts/csv/prices/BR/PETR3.SA.csv')
df_bench = df_bench.set_index('Date')
df_bench = df_bench[start_date:end_date]
df_bench['Daily Return'] = df_bench['Adj Close'].pct_change()
df_bench['Equity'] = (1 + df_bench['Daily Return']).cumprod() * initial_equity
df_bench = df_bench[['Equity']]
df_bench.rename(columns={'Equity':'PETR3'}, inplace=True)

for file in  br_files_list:
    if(file == 'FIBR3.SA.csv'):
        print('aqui')
    try:
        df = pd.read_csv(f'/home/jm/Projetos/Pessoais/stock-market/stock-market-scraper/py_scraper_scripts/csv/prices/BR/{file}')
        df = df.set_index('Date')
        df = df[start_date:end_date]
        df['Daily Return'] = df['Adj Close'].pct_change()
        df['Equity'] = (1 + df['Daily Return']).cumprod() * initial_equity
        df = df[['Equity']]
        column_name = file.replace('.SA', '').replace('.csv', '')
        df.rename(columns={'Equity':column_name}, inplace=True)
        df_bench = df_bench.join(df)
        if(str(df_bench[column_name].tail(1)[0]) == 'nan'):
            df_bench.drop(columns=[column_name], inplace=True)
    except:
        continue
```

```python
stock_equity = pd.read_csv('/home/jm/Projetos/Pessoais/stock-market/stock-market-scraper/py_scraper_scripts/csv/equit_curves.csv')
stock_equity = stock_equity.set_index('Date')
stock_equity
```

```python
stock_equity.T
```

```python
normalizer = Normalizer()
kmeans = KMeans(n_clusters=60)
pipeline = make_pipeline(normalizer, kmeans)

# Fit pipeline to the daily price movements
pipeline.fit(stock_equity.T)
labels = pipeline.predict(stock_equity.T)
len(labels)
```

```python
companies = stock_equity.columns.values
len(companies)
```

```python
# Create a DataFrame aligning labels and companies: df
df_labels = pd.DataFrame({'labels': labels, 'companies': companies})

# Display df sorted by cluster label
df_labels.groupby(by=['labels']).count()
```

```python
df_quant = pd.read_csv('/home/jm/Projetos/Pessoais/stock-market/stock-market-scraper/py_scraper_scripts/csv/df_quant_data_stock.csv')
df_quant_filter = df_quant[df_quant['MARKET'] == 'BR']
df_quant_filter = df_quant_filter[(df_quant['volume'] > 100000) & (df_quant['sharpe_1m'] > 2)]
df_quant_filter = df_quant_filter.sort_values(by=['sharpe_1y'], ascending=False)
df_quant_filter
```

```python
best_labels_cias = []

for _, row in df_quant_filter[['ticker']].head(10).iterrows():
    try:
        filter_symbol_best_sharpe_1y = row['ticker'].replace('.SA', '')
        
        # Use loc for cleaner indexing
        best_label_row = df_labels.loc[df_labels['companies'] == filter_symbol_best_sharpe_1y]
        
        # Check if the label is found
        if not best_label_row.empty:
            best_label = best_label_row.iloc[0, 0]
            
            matching_labels = df_labels[df_labels['labels'] == best_label]
            
            # Use append() for better performance
            for value in matching_labels.values:
                best_labels_cias.append(value)
            
    except Exception as e:  # Catch the exception and print it
        print(f"Error for ticker {row['ticker']}: {e}")
        continue

best_labels_cias = pd.DataFrame(columns=['label', 'cia'], data=best_labels_cias)
best_labels_cias = best_labels_cias.drop_duplicates(subset=['label'])
best_labels_cias
```
